```python
import numpy as np 
import pandas as pd 
```

```python
monthly_sales = pd.read_csv("monthly_sales.csv")
monthly_sales.head()
```

```python
monthly_sales['Month, Year'].to_numpy()
```

```python
for i in range(len(monthly_sales.columns)):
    print(monthly_sales.columns[i])
    print(monthly_sales.iloc[:,i].to_list())
    print()
```

```python
# total sales from each region by region
monthly_sales.sum(axis=0).to_numpy()
```

```python
# total sales from each region by month
monthly_sales.drop(['Month, Year'], axis=1).sum(axis=1).to_numpy()
```

```python
# get the difference in sales from the previous month 
diffs = monthly_sales.drop(['Month, Year'], axis=1).sum(axis=1).diff().fillna(0)
((diffs - diffs.mean()) / diffs.std()).to_numpy()
```
